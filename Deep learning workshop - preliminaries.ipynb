# **Creating Tensors** 

```python
x = tf.constant(4)
print(x)
```

```python
x.ndim
```

```python
# Creating matrix
x = tf.constant([[2, 4, 9],[1, 5, 3]], dtype=tf.float32)
print(x)
```

```python
x = tf.range(6, dtype=tf.float32)
print(x)
```

```python
# Reshaping
X = tf.reshape(x, (2, 3))
print(X)
```

```python
tf.range(start=2, limit=10, delta=2)
```

```python
# Changing dtype
x = tf.cast(x, dtype=tf.float64)
print(x)
```

```python
# Initializing tensors to contain all zeros or ones
x = tf.zeros((2, 3, 4)) 
print(x)
y = tf.ones((4, 4))
print(y)
```

```python
# Identity matrix
x = tf.eye(4)
```

```python
print(x)
```

```python
# Sampling from a ditribution
tf.random.normal(shape=[3, 5], mean=0, stddev=1)
```

```python
tf.random.uniform((1,3), minval=0, maxval=1)
```

# **Indexing and Slicing**

```python
X = tf.constant([2, 1, 4, 3, 5, 7, 8])
X[1:3]
```

```python
print(X[::2])
```

```python
print(X[::-1])
```

# **Mathematical Operations**

```python
x = tf.constant([1.0, 2, 4, 8])
y = tf.constant([2.0, 2, 2, 2])
```

```python
x + y
# tf.add(x, y)
```

```python
tf.add(x, y)
```

```python
x - y
# tf.subtract(x, y)
```

```python
x / y
# tf.divide(x, y)
```

```python
x * y
# tf.multiply
```

```python
tf.tensordot(x, y, axes=1)
```

```python
c = tf.range(9)
c = tf.reshape(c, (3,3))
print(c)
```

```python
tf.transpose(c)
```
